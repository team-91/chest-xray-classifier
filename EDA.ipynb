```python
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import pydicom
from pydicom.pixels import apply_voi_lut

import kagglehub
```

```python
def read_dicom(path):
    """
    Читает DICOM файл и конвертирует в numpy array
    """
    dicom = pydicom.dcmread(path)
    data = dicom.pixel_array.astype(np.float32)
    
    # Применяем масштабирование по тегам
    slope = float(dicom.get("RescaleSlope", 1.0))
    intercept = float(dicom.get("RescaleIntercept", 0.0))
    data = data * slope + intercept

    # Проверяем инверсию
    if dicom.PhotometricInterpretation == "MONOCHROME1":
        data = np.max(data) - data

    # Нормализация в 0-255
    data = data - np.min(data)
    data = data / np.max(data)
    data = (data * 255).astype(np.uint8)
    
    # RGB
    data_rgb = np.stack([data]*3, axis=-1)
    
    return data_rgb


def plot_single_dicom(image, title="", figsize=(8, 8)):
    """
    Визуализация одного DICOM изображения
    """
    plt.figure(figsize=figsize)
    plt.imshow(image)
    plt.title(title, fontweight='bold')
    plt.axis('off')
    plt.show()
```

## Корреляция между классами

> Наблюдается множество значимых корреляций между заболеваниями - особенно между плевральными, фиброзными и сердечно-сосудистыми патологиями

```python
class_presence = train_df.pivot_table(
    index='image_id', 
    columns='class_name', 
    values='class_id', 
    aggfunc='count', 
    fill_value=0
)
class_presence = (class_presence > 0).astype(int)
```

| Классы | Коэффициент корреляции |
|----------|---------------------|
| Cardiomegaly, Aortic enlargement | 0.69 |
| Pleural thickening, Pleural effusion | 0.55 |
| Pleural thickening, Pulmonary fibrosis  | 0.48 |
| Pleural thickening, Aortic enlargement  | 0.46 |
| Lung opacity, Pleural effusion | 0.42 |
| Lung opacity, Consolidation | 0.41 |
| Lung opacity, Pulmonary fibrosis | 0.40 |
| Infiltration, Lung opacity | 0.38 |
| Infiltration, Pulmonary fibrosis| 0.38 |
| Pleural thickening, Lung opacity| 0.37 |

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

class_corr = class_presence.corr()
mask = np.zeros_like(class_corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

cmap = sns.color_palette("coolwarm", as_cmap=True)

sns.heatmap(class_corr,
            annot=True,
            fmt='.2f',
            square=True,
            mask=mask,
            linewidth=3,
            vmax=0.5, vmin=-0.5,
            cmap=cmap,
            cbar=True,
            ax=ax)

ax.set_yticklabels(ax.get_xticklabels(), rotation = 0, fontsize=11)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=11)

plt.tight_layout()
plt.show()
```

## Стратегии борьбы с дисбалансом классов
- **Weighted Loss:**  
   Использование взвешенной функции потерь;
- **Oversampling:**  
   Увеличение числа примеров редких классов;
- **Data Augmentation:**  
   Искусственное расширение обучающего набора данных.

> Для классических ML-моделей планируется использовать **oversampling** 

> Для DL-моделей будем применять **data augmentation** и **weighted loss** 

### Метрики
Для мультиклассовой классификации с дисбалансом мы будем использовать:
- **Recall** - за основу берем такую метрику, так как опаснее пропустить патологию, чем ошибочно ее предсказать;

- **F1-score (macro + weighted)** - балансирует точность и полноту, macro-F1 позволяет оценить редкие патологии;

- **PR-AUC** - показывает способность модели выделять редкие заболевания при большом количестве "No finding";
- **mAP (mean Average Precision)** - средняя точность по всем классам.

### Корреляции между патологиями

>Анализ показал, что некоторые патологии часто встречаются вместе:

- Cardiomegaly и Aortic enlargement (0.69)

- Pleural thickening и Pleural effusion (0.55)

- Pleural thickening и Pulmonary fibrosis (0.48) и др.

>Практическое использование:

- **Feature engineering** для классических ML-моделей: создавать бинарные признаки "совместное наличие патологии";

- **Постобработка DL-предсказаний**: корректировка вероятностей редких, но коррелирующих классов (например, если модель уверена в Cardiomegaly, слегка повышаем вероятность Aortic enlargement).
